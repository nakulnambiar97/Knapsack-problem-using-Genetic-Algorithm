```python
import random as rand
import numpy as np
```

```python
def weight_calc(a,b,c):
    weight=np.dot(a,b)
    if weight<=c:
        return weight
    return 0
```

```python
weight_list=[50, 280, 84, 50, 80, 70, 21, 50, 240, 120, 36, 300, 120, 250, 75, 650, 125, 280, 84, 50, 145, 140, 42, 150]
num_pop=10
mutationrate=0.1
numelites=1
Numberofgenerations=200
popsize=len(weight_list)
max_wt=1500
```

```python
pop=[]
for h in range(num_pop):
    c=[]
    for i in range(popsize):
        a=np.random.randint(0,2)
        c.append(a)
    pop.append(c)
```

```python
for x in range(Numberofgenerations):
    weightage=[]
    fitnessmax=0
    for i in range(len(pop)):
        d=weight_calc(pop[i],weight_list,max_wt)
        weightage.append(d)
    mating_pool=rand.choices(pop,k=num_pop,weights=weightage)
    #extra
    fitnessmax=max(weightage)
    if fitnessmax==0:
        pop=[]
        for h in range(num_pop):
            c=[]
            for i in range(popsize):
                a=np.random.randint(0,2)
                c.append(a)
            pop.append(c)
        print("initial population",pop)
        continue
    #extra
    popweight=dict(zip(weightage,pop))
    weightage.sort(reverse=True)
    sortedpop=[]
    for i in weightage:
        sortedpop.append(popweight[i]) 
    #crossover
    pop=[]
    #pop.append(hu)
    for i in range(numelites):
        pop.append(sortedpop[i]) 
    while len(pop)<=num_pop:
        a1=rand.randint(0,num_pop-1)
        a2=rand.randint(0,num_pop-1)
        str1=mating_pool[a1]
        str2=mating_pool[a2]
        pos=rand.randint(1,len(str1))
        child1=str1[:pos]+str2[pos:]
        child2=str2[:pos]+str1[pos:]
        pop.append(child1)
        pop.append(child2)
    for i in range(numelites,num_pop):
        for j in range(popsize):
            n=rand.random()
            if n<mutationrate:
                if pop[i][j]==0:
                    pop[i][j]=1
                else:
                    pop[i][j]=0
    wr1=np.max(weightage)
    if np.average(weightage)!=0:
        print(x+1)
        print('Maximum Weight=', max(weightage))
        #print(max(weightage),"Max weight = ",wr1,x+1,pop[0])
    print(weightage)
    print()
```

```python
knapsack_solution_weights=[]
for i in range(popsize):
    if pop[0][i]==1:
        knapsack_solution_weights.append(weight_list[i])
print(knapsack_solution_weights, 'total weight = ',sum(knapsack_solution_weights))
```
